## 6.1. 構造主義とは
- 思想なり哲学で必ず出てくるのが、へーゲル、カント、ニーチェ、フッサール、メルローポンティといった哲学者やマルクスの資本論がでてきます。さらに構造主義、ポスト構造主義も出てきます。構造主義とはなにか、構造とはなにか　ってのは思想としての構造主義だととらえようがない。[構造主義とは](https://liberal-arts-guide.com/structuralism/#:~:text=%E6%A7%8B%E9%80%A0%E4%B8%BB%E7%BE%A9%EF%BC%88structuralism%EF%BC%89%E3%81%A8%E3%81%AF,%E3%81%8B%E3%82%89%E3%81%8D%E3%81%A6%E3%81%84%E3%82%8B%E3%81%AE%E3%81%A7%E3%81%99%E3%80%82) の中で紹介されてる本「はじめての構造主義」という本から見ていきます。この本は比較的読みやすい本のように思います。有名な「構造と力」はなかなか読んでもピンときません。その一つの原因として、「はじめての構造主義」やWikiではきちんと数学との関係が言われてますが、数学における数学的構造を踏まえて書いてるか、ないしはその数学的内容をどれだけ理解しているかの度合いによって、書いてある内容がわかりやすかったり、わかりにくかったりするような気がします。

- クロード・レヴィ＝ストロースが構造主義の生みの親となりますが、彼は、ヤーコブソンから自分の問題意識への解決するための方法論、それがソシュールの言語学、これが構造主義言語学というべきか、その具体的な解決のためのある部族の婚姻関係のルールの解明を実は数学者に頼んでる。その数学者こそ、ブルバキに属していた有名な数学者、アンドレ・ヴェイユで、かれが結局、クライン群という代数的構造で解明した、結局、構造主義のルーツはこのヴェイユが使った代数的構造でした。

- これまで、数についてから、数学的構造で代数的構造や位相構造、順序構造などにふれてきました。ほかにも多様体構造などいろいろあるわけですが、一般の思想の構造主義はヴェイユが示した比較的単純な代数的構造のなかでの群構造、特にクライン群の構造だったのですが、そこから独自に発展して、「構造」という言葉が出来上がっていったような感じです。サルトルとかの不決定の自由のための実存主義に対して、どうしても変えられない普遍的な法則性のようなもの、例えば、これは違うかもしれませんが、実存主義に対して、フロイトの無意識の存在は、意識化にない人間を支配するもの、それは今日の私たちの脳の働きを知る上でも重要になってきてますが、そういうものによって支配される。あるいは、言語学はソシュールとは別に、後期ウィトゲンシュタインの思想は、言語によって哲学とりわけ形而上学が支配されそれを言語哲学によって明らかにしようとしたのが前期で「論考」だが、それを後期は自ら壊していく、それはフッサールの言語哲学において、ゲーデルの不完全性定理とっいった言語の持つ深い淵、そのあたりが差異の体系としてソシュールの言語学と関係しそうですが。

- 一方で、既存の体制を否定して、それとは違う「構造」があるという考え方、構造主義にあるみたいで、これが反体制的な運動の思想的根拠として、構造主義が一つのムーブメントやブームになった部分でないかと思います。これはあくまで個人的な私論です。このあたりになってくると、構造という言葉が、独り歩きして、しかもただバズワードなってしまった感があります。

- ソシュールの差異の体系にもあるように、実体概念ではなく関係概念が重要で、マルティン・ブーバーの「我と汝」の「初めに関係ありき」に通じるものあるのかも、そしてそれはブルバキの構造主義的な数学は自然に圏論をふんだんに使った今日的な現代数学へと自然に流れてい行くのもうなずけるような気がします。

- このように言語学と構造主義は切っても切れない関係があり、そこにはソシュールや後期ウィトゲンシュタインの思想。この思想をいろいろ語るのは台湾のあのオードリー・タン氏です。彼女は哲学に詳しい。言語のもつ深遠さが今日のディープラーニングによる機械学習によってはじめて自然言語処理やAIができるのは、それまでのロジック重視の人工知能開発がフッサールの言語哲学のようなあるいは前期ウィトゲンシュタインのような、ある意味、言語の完全性という幻想のなかで解決案を求めようとしたからだと書いてたと思います。
